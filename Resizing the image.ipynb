```python
import cv2, numpy as np, matplotlib.pyplot as plt, matplotlib.image as img_
```

```python
img= img_.imread(path)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
gaussian_blur_mat= np.array([[1,2,1],[2,4,2],[1,2,1]])*(1/16)
```

```python
blurred_img=cv2.filter2D(img, -1, kernel=gaussian_blur_mat)
```

```python
plt.imshow(blurred_img)
```

### Detecting vertical and horizontal edges using Sobel Kernel

```python
sobel= np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
sobel_1=np.transpose(sobel)
```

```python
sobel_x= cv2.filter2D(img, -1, sobel) 
```

```python
plt.imshow(sobel_x)
```

```python
sobel_y= cv2.filter2D(img, -1, sobel_1) 
```

```python
plt.imshow(sobel_y)
```

```python
plt.imshow(cv2.Sobel(img,cv2.CV_64F, 1,0, ksize=7))
```

```python
plt.imshow(cv2.Sobel(img,cv2.CV_64F, 0,1, ksize=7))
```

### Translating image

```python
translate=np.float32([[1,0,10],
                    [0,1,20]])
```

```python
trans=cv2.warpAffine(img,translate,(img.shape[0],img.shape[1]))
```

```python
img.shape[0]
```

```python
img.shape[1]
```

```python
plt.imshow(trans, cmap= "viridis")
```

### Rotating images

```python
height=img.shape[0]
width=img.shape[1]
matrix= cv2.getRotationMatrix2D((height/2,width/2),90,1)
```

```python
rotated_image=cv2.warpAffine(img,matrix,(width,height))
```

```python
plt.imshow(rotated_image)
```
